```python
#summary of all the columns

def check (df):
    l=[]
    columns=df.columns
    for col in columns:
        dtypes=df[col].dtypes
        nunique= df[col].nunique()
        sum_null=df[col].isnull().sum()
        l.append([col,dtypes,nunique,sum_null])
    df_check=pd.DataFrame(l)
    df_check.columns=["column", "dtypes", "nunique","sum_null"]
    return df_check
check(netflix)
```

```python
#movie type by country
movie_type_by_relyear = pd.crosstab(netflix["type"], netflix["release_year"].nunique())
movie_type_by_relyear
```

```python
# movie type by months added
movie_by_month = pd.crosstab(netflix["type"], netflix["month"])
movie_by_month
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(data= netflix, x='month', hue = 'type', palette='Set1')
plt.title('Count of Movies and TV Shows Added by months')
plt.xlabel('Month added')
plt.ylabel('type')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# movie type by year added
movie_by_year = netflix.groupby("year_added")["type"].count()
movie_by_topyear = movie_by_year.sort_values(ascending=False).head(5)
movie_by_topyear
```

```python
plt.figure(figsize=(6, 4))
movie_by_topyear.plot(kind='line', x='year_added', y='count', label='Movie Count by Year')
plt.xlabel('Year_added')
plt.ylabel('Count')
plt.title('Movie Count by Year')
plt.show()
```

```python
plt.figure(figsize =(15,9))
sns.countplot(x = "rating", data = netflix, hue = "type", palette='Set2')
plt.xticks(rotation = 45)
plt.title("Movies rating by type", fontdict= {"fontsize": 20})
plt.show()
```

```python
# Calculating the top 5 country based on type of movies

type_country = netflix.groupby("country")["type"].count()
top_5_country = type_country.sort_values(ascending=False).head(5)

# Printing the top 5 cities by sales
print(top_5_country)
```

```python
top_5_country.plot.bar()

# Adding a legend to the plot
plt.legend(loc='upper right')

# Adding a title to the plot
plt.title(" Top 5 Country most watched movie type", fontdict={"fontsize": 20})
```

```python
# movies types
fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(data=netflix['type'].value_counts(), x = netflix['type'].value_counts(), labels=netflix['type'].unique(), colors= ['blue', 'green', '#FF5733'], radius=3, center=(4, 4), wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)

# Show the plot
plt.show()
```

```python
# Calculating the least 5 country based on type of movies

type_country = netflix.groupby("country")["type"].count()
least_5_country = type_country.sort_values(ascending=False).tail(5)

# Printing the top 5 cities by sales
print(least_5_country)
```
